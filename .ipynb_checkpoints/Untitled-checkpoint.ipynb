```python
import pandas as pd
```

```python
import numpy as np
```

```python
df1 = pd.read_csv('data1.csv')
df1
```

```python
df1['static_bias_X'] = df1['OmegaX_[deg/s]'].mean()
df1['static_bias_Y'] = df1['OmegaY_[deg/s]'].mean()
df1['static_bias_Z'] = df1['OmegaZ_[deg/s]'].mean()

df1['static_acc_X'] = df1['AccX_[m/s/s]'].mean()
df1['static_acc_Y'] = df1['AccY_[m/s/s]'].mean() + 1.1
df1['static_acc_Z'] = df1['AccZ_[m/s/s]'].mean() - 9.81

df1
```

```python
df2 = pd.read_csv('data2.csv')
df2
```

```python
df2['static_bias_X'] = df2['OmegaX_[deg/s]'].mean()
df2['static_bias_Y'] = df2['OmegaY_[deg/s]'].mean()
df2['static_bias_Z'] = df2['OmegaZ_[deg/s]'].mean()

df2['static_acc_X'] = df2['AccX_[m/s/s]'].mean()
df2['static_acc_Y'] = df2['AccY_[m/s/s]'].mean() + 1.1
df2['static_acc_Z'] = df2['AccZ_[m/s/s]'].mean() - 9.81

df2
```

```python
df3 = pd.read_csv('data3.csv')
df3
```

```python
df3['static_bias_X'] = df3['OmegaX_[deg/s]'].mean()
df3['static_bias_Y'] = df3['OmegaY_[deg/s]'].mean()
df3['static_bias_Z'] = df3['OmegaZ_[deg/s]'].mean()

df3['static_acc_X'] = df3['AccX_[m/s/s]'].mean()
df3['static_acc_Y'] = df3['AccY_[m/s/s]'].mean() + .1
df3['static_acc_Z'] = df3['AccZ_[m/s/s]'].mean() - 9.81

df3
```

```python
df_full = pd.read_csv('data4.csv')
df_full
```

```python
staticX = (df1.static_bias_X[0] + df2.static_bias_X[0] + df3.static_bias_X[0]) / 3
staticY = (df1.static_bias_Y[0] + df2.static_bias_Y[0] + df3.static_bias_Y[0]) / 3
staticZ = (df1.static_bias_Z[0] + df2.static_bias_Z[0] + df3.static_bias_Z[0]) / 3

r2rX = np.ptp([df1.static_bias_X[0], df2.static_bias_X[0], df3.static_bias_X[0]])
r2rY = np.ptp([df1.static_bias_Y[0], df2.static_bias_Y[0], df3.static_bias_Y[0]])
r2rZ = np.ptp([df1.static_bias_Z[0], df2.static_bias_Z[0], df3.static_bias_Z[0]])

inrX = np.sqrt(np.mean((df_full['OmegaX_[deg/s]'])-staticX)**2)
inrY = np.sqrt(np.mean((df_full['OmegaY_[deg/s]'])-staticY)**2) 
inrZ = np.sqrt(np.mean((df_full['OmegaZ_[deg/s]'])-staticZ)**2)

staticXa = (df1.static_acc_X[0] + df2.static_acc_X[0] + df3.static_acc_X[0]) / 3
staticYa = (df1.static_acc_Y[0] + df2.static_acc_Y[0] + df3.static_acc_Y[0]) / 3
staticZa = (df1.static_acc_Z[0] + df2.static_acc_Z[0] + df3.static_acc_Z[0]) / 3

r2rXa = max(df1.static_acc_X[0], df2.static_acc_X[0],  df3.static_acc_X[0]) - min(df1.static_acc_X[0], df2.static_acc_X[0],  df3.static_acc_X[0])
r2rYa = max(df1.static_acc_Y[0], df2.static_acc_Y[0],  df3.static_acc_Y[0]) - min(df1.static_acc_Y[0], df2.static_acc_Y[0],  df3.static_acc_Y[0])
r2rZa = max(df1.static_acc_Z[0], df2.static_acc_Z[0],  df3.static_acc_Z[0]) - min(df1.static_acc_Z[0], df2.static_acc_Z[0],  df3.static_acc_Z[0])

inrXa = np.sqrt(np.mean(sum(df_full['AccX_[m/s/s]'])-staticX)**2) / len(df_full)
inrYa = np.sqrt(np.mean(sum(df_full['AccY_[m/s/s]'])-staticY)**2) / len(df_full)
inrZa = np.sqrt(np.mean(sum(df_full['AccZ_[m/s/s]'])-staticZ)**2) / len(df_full)


data = pd.DataFrame({'Static': [staticX, staticY, staticZ,staticXa, staticYa, staticZa] , 'R2R' : [r2rX, r2rY, r2rZ,r2rXa, r2rYa, r2rZa], 'InRun' : [inrX, inrY, (inrZ),inrXa, inrYa , (inrZa- 9.81)]})
data
```
